```python
#importing needed modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
#choose plots style
sns.set_style('darkgrid')
#make sure plots are inline with the notebook
%matplotlib inline
```

```python
## Loading the dataset and checking the columns we have

### Load your data and print out a few lines. Perform operations to inspect data
### Types and look for instances of missing or possibly errant data.
data_frame = pd.read_csv('no_show_appointments.txt')
data_frame.head()
```

```python
### Get the shape and types of our data
print(data_frame.shape)
pd.DataFrame(data_frame.dtypes)
```

```python
### Get some statistics about our data
data_frame.describe()
```

```python
### Check if there is any missing values in our data
data_frame.info()
data_frame.isna().any()
```

```python
### Check if there is any duplicated rows in our data
data_frame.duplicated().any()
```

```python
### Drop irrelevant columns
data_frame.drop(['PatientId','AppointmentID'],axis=1,inplace=True)
data_frame.head()
```

```python
### Change all cloumns name to lower case and replace all - with _
data_frame.columns=data_frame.columns.str.lower().str.replace('-','_')
pd.DataFrame(data_frame.columns)
```

```python
### Change data columns to data type
data_frame['scheduledday']=pd.to_datetime(data_frame['scheduledday'])
data_frame['appointmentday']=pd.to_datetime(data_frame['appointmentday'])
```

```python
### Turn no_show column to show, no&yes to 0&1
print(data_frame.no_show.unique())
data_frame.no_show=data_frame.no_show.map({'No':1,'Yes':0})
data_frame.rename(columns={'no_show':'show'},inplace=True)
print(data_frame.show.unique())
data_frame.head()
```

```python
### Create a new column for days difference between scheduling and appointment
day_diff=(data_frame.appointmentday.dt.date-data_frame.scheduledday.dt.date).dt.days
data_frame.insert(3,'day_diff',day_diff)
data_frame.day_diff.dtype
```

```python
### Check data one last time
data_frame.dtypes
```

```python
#define function to get the ratio of show in different categories
def plot_rat(x):
    data_frame.groupby(x).show.mean().plot(kind='bar',
                                    edgecolor='black',
                                    figsize=(14,8)).set_ylabel('Ratio of show');
    display(data_frame.groupby(x)[['show']].mean())
    plt.legend()
```

```python
#get some statistics about our data
data_frame.describe()
```

# What is the percentage of patients who did not show up?

```python
# percentage of show and no show
print(f"percentage of patients who didn't show up for their appointment is { (1-data_frame.show.mean())*100 } %" )
no_show=len(data_frame[data_frame.show==0])/len(data_frame.show)
show=len(data_frame[data_frame.show==1])/len(data_frame.show)
plt.bar(['show','no show'],[show*100,no_show*100],color=['g','r']);
plt.title('Percentage of show and no_show');
plt.ylabel('Percentage');
plt.xlabel('show or no-show');
display(data_frame.groupby('show')[['show']].count())
```

```python
#create filters for show and no-show
show=(data_frame.show == 1)
no_show=(data_frame.show == 0)
total_miss=len(data_frame[no_show])
total=len(data_frame)
```

```python
#get the number of patients missing their appointments by gender
no_show_gender=data_frame[no_show]['gender'].value_counts()
no_show_gender.plot(kind='pie');
plt.title('Patients who missed their appointment by gender');
print('percentage of Females and Males who missed their appointment:')
#get the percentage of patients missing their appointments by gender
pd.DataFrame(no_show_gender*100/total)
```

```python
data_frame.groupby(['gender','show']).size().unstack('gender').plot(kind='bar').set_ylabel('number of patients')
```

```python
#what is the percentage of patients missing their appointment by scholarship
plot_rat(data_frame.scholarship)
plt.title('Ratio of show or no-show by scholarship');
```

```python
#what is the percentage of patient who attended their appointment by sms_received
plot_rat(data_frame.sms_received)
plt.title('Ratio of show or no-show by sms_received');
```

```python
#filter for positive day difference
df1=data_frame[data_frame.day_diff>=0]
# df1.day_diff.unique()
#turn day diff into categorical column Day_diff2
bin_edges=[-1,0,4,15,179]
names=['sameday','fewdays','more_than_4','more_than_15']
data_frame['day_diff2']=pd.cut(df1.day_diff,bin_edges,labels=names)
#filter for no-show records and count values for each category of day_diff2
```

```python
no_show_day_diff=data_frame[no_show].day_diff2.value_counts()/len(data_frame[no_show])*100
no_show_day_diff.reindex(names).plot(kind='bar');
plt.title('propotion of time difference for no-show appointments');
plt.xlabel('amount of days between scheduling and appointment');
plt.ylabel('Ratio of no-show');
print('The propotion of time difference for no-show appiontments:')
pd.DataFrame(no_show_day_diff)
```

# The longer the time between schedule and appointment day, the higher the chance of missing appointment
